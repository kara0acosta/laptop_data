## T Tests

```python
# This code performs a statistics operation
gaming = df[df["Type"] == "Gaming"]["Price (USD)"]
notebook = df[df["Type"] == "Notebook"]["Price (USD)"]

st.ttest_ind(gaming, notebook, equal_var=False)
```

## Confidence Interval

```python
# This code performs a statistics operation
confidence = 0.95
data = df["Price (USD)"]

mean = np.mean(data)
sem = st.sem(data)

interval = st.t.interval(confidence, len(data)-1, loc=mean, scale=sem)

interval
```

## ANOVA

## Linear Regression

```python
# This code performs a statistics operation
x = df["RAM_GB"]
y = df["Price (USD)"]

slope, intercept, r, p, std_err = st.linregress(x, y)

def line(x):
    return slope * x + intercept

model = list(map(line, x))

plt.scatter(x, y)
plt.plot(x, model)
plt.title("Linear Regression")
plt.show()
```

```python
# This code performs a statistics operation
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Sample Dataset
df = np.array([85, 90, 78, 92, 88, 76, 95, 89, 82, 80])

mean_val = np.mean(df)
std_val = np.std(df)

print(f'Mean: {mean_val:.2f}')
print(f'Standard Deviation: {std_val:.2f}')
```

```python
# Z-scores for all df points
z_scores = stats.zscore(df)
print(f'Z-scores: {z_scores}')

# Finding X from a Z-score (e.g., Z = 1.2)
z_input = 1.2
x_val = mean_val + (z_input * std_val)
print(f'Raw score (X) for Z=1.2: {x_val:.2f}')
```

```python
# This code performs a statistics operation
normal_dist = np.random.normal(50, 10, 1000)
skewed_dist = stats.skewnorm.rvs(a=10, loc=50, scale=10, size=1000)

plt.figure(figsize=(10, 5))
plt.hist(normal_dist, bins=30, alpha=0.5, label='Normal Distribution', color='skyblue')
plt.hist(skewed_dist, bins=30, alpha=0.5, label='Right Skewed', color='salmon')
plt.title("Visualizing Skewness")
plt.legend()
plt.show()
```

```python
# Probability of a score being less than 80
prob = stats.norm.cdf(80, loc=mean_val, scale=std_val)
print(f'Probability (X < 80): {prob:.4f}')

# T-test p-value (Testing if the df mean differs from 85)
t_stat, p_val = stats.ttest_1samp(df, 85)
print(f'P-Value: {p_val:.4f}')
```

```python
# This code performs a statistics operation
q1, q2, q3 = np.percentile(df, [25, 50, 75])
print(f'Q1: {q1}, Median: {q2}, Q3: {q3}')

plt.figure(figsize=(7, 5))
plt.boxplot(df, vert=False, patch_artist=True,
            boxprops=dict(facecolor='lightgreen', color='black'))
plt.title("Box-and-Whisker Plot")
plt.xlabel("Scores")
plt.show()
```

```python
# This code performs a statistics operation
import numpy as np
from scipy import stats

# Dataset representing scores
df = np.array([85, 90, 78, 92, 88, 76, 95, 89, 82, 80])

# --- T-test Calculation ---
# Null Hypothesis: The mean of the df is equal to 85
null_mean = 85
t_statistic, p_value = stats.ttest_1samp(df, null_mean)

print(f"--- T-test Analysis ---")
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# --- Decision Making ---
alpha = 0.05
if p_value < alpha:
    print("Decision: Reject the Null Hypothesis (Significant difference found).")
else:
    print("Decision: Fail to Reject the Null Hypothesis (No significant difference found).")
```
